# Gradient descent

```python
x_mean = np.mean(x)
x_std = np.std(x)
x = (x - np.mean(x)) / np.std(x)
y = (y - np.mean(y)) / np.std(y)
x_p = x.reshape(1, x.size)
```

```python
x
```

```python
y
```

```python
# my model:
# y = w0 + w1*x

from jax import grad
import jax.numpy as jnp

def model(x, w):
    a = w[0] + jnp.dot(x.T, w[1:])
    return a.T

def least_squares(w):
    global x_p, y
    cost = jnp.sum((model(x_p,w) -y)**2)
    return cost/float(y.size)

gradient = grad(least_squares)


# this function is defined to check whether jax.grad can give me the correct result
def gradient_test(w):
    global x, y
    x0 = np.ones(len(x))
    x_prime = np.vstack([x0, np.array(x)]).T
    y_pred = np.matmul(x_prime, np.array(w))

    w0_grad = 2 * np.sum((y_pred - y)  ) / len(x)
    w1_grad = 2 * np.sum(np.dot((y_pred - y), x.T ) )/ len(x)
    return np.array([w0_grad, w1_grad])
```

```python
w = np.random.rand(2)
print(w)
```

```python
# testing gradient function
print("grad of jax.grad: ", gradient(w))

print("grad of self-defined gradient: ", gradient_test(w))
```

```python
w = np.array([1., 1.])  # np.random.rand(2)
cost = least_squares(w)
w_history = [w,]
cost_history = [cost,]
alpha = np.array([1e-2, 1e-2]) # learning rate for w0 and w1
beta = 0.9
beta_2 = 0.999
epsilon = 1e-8
grad_history = [gradient(w), ]
momentum_history = [np.zeros(2), ]
secondary_list = [np.zeros(2), ]

for i in np.arange(1,200,1):
    # get gradient
    w_grad = gradient(w)
    grad_history.append(w_grad)
    
    # get momentum
    momentum = (1-beta)*(w_grad) + beta*momentum_history[-1]
    momentum_history.append(momentum)
    
    # second derivative estimate
    w_grad_square = w_grad**2
    secondary = beta_2 *secondary_list[-1] + (1-beta_2) *w_grad_square
    secondary_list.append(secondary)
    
    # momentum with normalization
    w = w - ( (alpha * momentum / (1-jnp.power(beta, i)) )/ ( jnp.sqrt(secondary/(1-jnp.power(beta_2, i))) + epsilon) )
    w_history.append(w)

    # cost
    cost = least_squares(w)
    cost_history.append(cost)

```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,8))
axs[0,0].plot(cost_history, "ro-", label="cost")
axs[0,0].title.set_text('Cost')
axs[0,0].legend(loc="upper right")
# axs[0,0].set_ylim([0,10])


axs[0,1].plot(x, y, "ko", label="original data")
for idx,wi in enumerate(w_history[-1:]):
    axs[0,1].plot(x, model(x_p, wi), "o", label="predicted")
axs[0,1].title.set_text('Model prediction')
axs[0,1].legend(loc="upper right")

axs[1,0].plot([i[0] for i in grad_history], "go-", label="gradient of w0")
axs[1,0].plot([i[1] for i in grad_history], "yo-", label="gradient of w1")
axs[1,0].title.set_text('Gradients')
axs[1,0].legend(loc="upper right")
axs[1,0].set_ylim([-10,10])

w0, w1 = w_history[-1]
y_mean = np.mean(data[:,1])
y_std = np.std(data[:,1])
w0_original = (w0 - w1 * (x_mean / x_std)) * y_std + y_mean
w1_original = w1 * y_std / x_std

# the close solution
# A = np.vstack([x, np.ones(len(x))]).T
# w1_closed, w0_closed  = np.linalg.lstsq(A, y, rcond=None)[0]
# y_closed = model(x_p, np.array([w0_closed, w1_closed]))
axs[1,1].plot(data[:,0], data[:,1], "ko", label="original data")
axs[1,1].plot(data[:,0], model(data[:,0].reshape(1, x.size), np.array([w0_closed, w1_closed])), "go", label="closed solution")


print(w0_original, w1_original)

plt.show()
```
